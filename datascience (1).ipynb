```python
titanic_data.drop("Cabin",axis=1,inplace=True)
```

```python
titanic_data.head(5)
```

```python
titanic_data.dropna(inplace=True)
```

```python
sns.heatmap(titanic_data.isnull(), yticklabels=False)
plt.show()
```

```python
titanic_data.isnull().sum()
```

```python
#convert string variable into catagoricalvariable to implement logistic regression
pd.get_dummies(titanic_data['Sex'],dtype=int)
```

```python
sex=pd.get_dummies(titanic_data['Sex'],dtype=int,drop_first=True)
```

```python
embark=pd.get_dummies(titanic_data["Embarked"],dtype=int)
embark.head(5)
```

```python
embark=pd.get_dummies(titanic_data["Embarked"],dtype=int,drop_first=True)
embark.head(5)
```

```python
pcl=pd.get_dummies(titanic_data["Pclass"],dtype=int,drop_first=True)
pcl.head(5)
```

```python
titanic_data=pd.concat([titanic_data,sex,embark,pcl],axis=1)
```

```python
titanic_data.head(5)
```

```python
titanic_data.drop(['Sex','Embarked','PassengerId','Name','Ticket'],axis=1,inplace=True)
```

```python
titanic_data.drop(['Pclass'], axis=1, inplace=True)
```

```python
titanic_data.head(5)
```

## Train data

```python
X=titanic_data.drop("Survived",axis=1)
y=titanic_data["Survived"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from  sklearn.linear_model import LogisticRegression
```

```python
logmodel= LogisticRegression()
```

```python
X_train.columns = X_train.columns.astype(str)
X_test.columns = X_test.columns.astype(str)
```

```python
logmodel.fit(X_train,y_train)
```

```python
predictions=logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test,predictions)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,predictions)
```

```python
from sklearn.metrics import accuracy_score
```

```python
 accuracy_score(y_test,predictions)
```
